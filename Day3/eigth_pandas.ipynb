```python
import pandas as pd

df1 = pd.DataFrame({
    'city' : ['mumbai','delhi','chennai','banglore'],
    'temperature' : [32,45,33,34]
})

df1
```

```python
# creating another dataframe here
df2 = pd.DataFrame({
    'city' : ['mumbai','delhi','chennai','new delhi'],
    'humidity' : [23,45,23,14]
})

df2
```

```python
# merging both of these into a new dataframe
df3 = pd.merge(df1,df2,on = 'city')                     # [Inner Join]
df3
```

```python
# merging them together with outer join
df3 = pd.merge(df1,df2,on = 'city',how = 'outer',indicator = True)      # including all other ones can also be joined from right or left side
df3
```

```python
# adding a data containing different things together
df2 = pd.DataFrame({
    'city' : ['mumbai','delhi','chennai','banglore'],
    'temperature' : [32,45,33,34],
    'humidity' : [23,12,14,11]
})

# merging this new one with the first one
df3 = pd.merge(df1,df2,on = 'city',suffixes = ('_left','_right'))
df3
```
