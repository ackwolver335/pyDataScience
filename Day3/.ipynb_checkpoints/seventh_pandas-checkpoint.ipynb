```python
import pandas as pd

# creating a dataframe for Indian Cities weather
indian_weather = pd.DataFrame({
    'city' : ['mumbai','delhi','chennai','banglore'],
    'temperature' : [32,45,33,34],
    'humidity' : [80,60,45,70]
})

indian_weather
```

```python
# creating another dataframe for United States
us_weather = pd.DataFrame({
    'city' : ['NewYork','London','Berlin','Chicago'],
    'temperature' : [34,33,45,34],
    'humidity' : [60,45,33,89]
})

us_weather
```

```python
# concatenating two dataframe together
df = pd.concat([indian_weather,us_weather],ignore_index=True)
df
```
