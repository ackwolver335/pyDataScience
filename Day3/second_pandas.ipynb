```python
import pandas as pd

# first way to create a dataframe : Using CSV
df = pd.read_csv("weather-report.csv")
df
```

```python
# creating a dataframe with the help of excel file here
df1 = pd.read_excel("weather-report.xlsx","weather-report")
df1
```

```python
# Creating a dataframe using dictionary's object : 3rd Way
dict1 = {
    'date' : ['01/02/2024','01/03/2024','01/04/2024'],
    'temperature' : [34,45,66],
    'event' : ['Sunny','Hot','Pleasent']
}

df2 = pd.DataFrame(dict1)
df2
```

```python
# 4rth Way : Creating a DataFrame using list of tuples
tup1 = [
    ('01/02/2024',17,'Rainy'),
    ('01/03/2024',20,'Pleasent'),
    ('01/04/2024',30,'Sunny')
]

df3 = pd.DataFrame(tup1,columns = ['Date','Temperature','Event'])
df3
```

```python
# 5th Way : Creating a Dataframe using list of Dictionaries
list1 = [
    {'Date':'01/02/2024','Temperature':20,'Event':'Sunny'},
    {'Date':'01/03/2024','Temperature':15,'Event':'Rainy'},
    {'Date':'01/04/2024','Temperature':40,'Event':'Hot'}
]

df4 = pd.DataFrame(list1)
df4
```
