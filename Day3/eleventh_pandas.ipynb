```python
import pandas as pd
import numpy as np 

df = pd.read_excel("read_one.xlsx")
df
```

```python
# using the crosstab method
pd.crosstab([df.Nationality,df.Name],[df.Sex,df.Handedness],margins=True)              # margin for getting the total
```

```python
# using the normalize argument
pd.crosstab([df.Sex],[df.Handedness],values=df.Age,aggfunc=np.average)
```
