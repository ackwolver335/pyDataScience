```python
# replacing all the values of NaN seperately column-wise
df = pd.read_csv('weather-report.csv',index_col = False)
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'no event'
})

new_df
```

```python
# getting the use of forwordfill (ffill) and bfill (backward fill)
new_df = df.fillna(method="ffill")
new_df
```

```python
# using axis for making them towards the axis
new_df = df.fillna(method="ffill",axis = "columns")             # for horizontally copying them
new_df
```

```python
# using axis for making them towards the axis
df = pd.read_csv("weather-report.csv")
new_df = df.fillna(method="ffill",limit = 1)
new_df
```
