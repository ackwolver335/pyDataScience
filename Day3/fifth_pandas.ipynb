```python
import pandas as pd 
import numpy as np

df = pd.read_csv("weather-report-copy.csv")
df
```

```python
# replacing all the values that are negative
new_df = df.replace(-999,np.NaN)                # for replacing more values df.replace([list_of_values],replacable_value)
new_df
```

```python
# replacing values based on columns
new_df = df.replace({
    'temperature' : -999,
    'windspeed' : -999,
    'event' : 'No event'
},np.NaN)

new_df
```

```python
# replacing the values specifically with others
new_df = df.replace({
    -999 : np.NaN,
    'No event' : 'Normal Day'
})

new_df
```

```python
# replacing the units from the values
# new_df = df.replace('[A-Za-z]','',regex=True)                     # it also erased the one that we don't wanted to
new_df = df.replace({                                               # column based
    'temperature' : '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
},'',regex = True)
new_df
```

```python
# creating another dataframe here for conept of replacing the list of values from another list of values
df = pd.DataFrame({
    'score' : ['exceptional','average','good','poor','average','exceptional'],
    'student' : ['rob','manya','parthiv','tom','julian','erica']
})

df
```

```python
# replacing these values with the help of providing another list
new_df = df.replace(['exceptional','average','good','poor','average','exceptional'],[2,3,4,2,4,1])
new_df
```
