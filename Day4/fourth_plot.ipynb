```python
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
# creating the data to be displayed at the time plotting the bar chart
company = ['GOOGLE','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profitNB = [40,2,34,12]

# plotting the bar chart
ypos = np.arange(len(company))

# replacing the values now
plt.xticks(ypos,company)

# adding labels and titles
plt.title("US Tech Stacks")
plt.xlabel("Companies")
plt.ylabel("Revenue [US $]")

# now changing them to horizontal format
plt.barh(ypos+0.2,profitNB,label = 'Profit')
plt.barh(ypos-0.2,revenue,label = 'Revenue')
plt.legend()
```

```python
plt.bar(ypos+0.2,profitNB,width = 0.4,label = 'Profit')
plt.bar(ypos-0.2,revenue,width = 0.4,label = 'Revenue')
plt.legend(loc = 'best')
```
