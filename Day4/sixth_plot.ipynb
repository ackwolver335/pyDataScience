```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# getting the data ready
exp_values = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bills','Car','Other Utilities']

# plotting it
plt.axis('equal')                                   # for making it equally visualized
plt.pie(exp_values,labels = exp_labels,radius = 1,autopct = '%0.0f%%',shadow = True,explode = [0,0.2,0.2,0,0],startangle = 90)
plt.show()                                          # just for observing or getting the pie chart only
```
