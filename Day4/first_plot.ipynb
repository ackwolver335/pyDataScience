```python
import matplotlib.pyplot as plt

# data regarding the plot
x = [1,2,3,4,5]
y = [23,33,45,23,51]

# title of the Plot
plt.title("Weather Plots")

# labels regarding both the axis
plt.xlabel("Days")
plt.ylabel("Temperature")

# Plotting the chart here
plt.plot(x,y,color = 'red',linewidth = 2,linestyle = 'dotted')
```
