```python
import matplotlib.pyplot as plt
%matplotlib inline

# getting the data ready
days = [1,2,3,4,5]
max_t = [34,45,43,12,34]
min_t = [12,23,24,14,20]
avg_t = [35,37,40,42,39]

# adding labels and titles
plt.title("Temperature Visualization")
plt.xlabel("Days")
plt.ylabel("Temperature")

# plotting the data
plt.plot(days,max_t,label = "Max Temp")
plt.plot(days,min_t,label = "Min Temp")
plt.plot(days,avg_t,label = "Avg Temp")

# setting up the legend for showing particular legend
plt.legend(loc = 'best',shadow = True,fontsize = 'medium')           # loc argument for setting up the location of the legend and best for best position

# we can also enable the grid regarding the plot
# plt.grid()
```
