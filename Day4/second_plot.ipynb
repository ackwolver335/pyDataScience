```python
import matplotlib.pyplot as plt 
%matplotlib inline

# supplying the data here
x = [1,2,3,4,5,6]
y = [45,34,23,13,33,45]

# title of the plot
plt.title("Temperature Prediction")

# adding the labels
plt.xlabel("Days")
plt.ylabel("Temperature")

# plotting it to a chart
plt.plot(x,y,'rD')
```

```python
# another way of doing the same thing
plt.plot(x,y,color = '#292826',marker = 'D',linestyle = '')                 # markersize to increase the size of the marker
```

```python
# another way of doing the same thing
plt.plot(x,y,color = '#292826',marker = 'D',alpha = 0.2)                # alpha for getting the marker ready
```
